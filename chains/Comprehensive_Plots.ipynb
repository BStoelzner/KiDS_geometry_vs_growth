```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('classic')
%matplotlib inline

def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))

fig, ax = plt.subplots()
x_values = np.linspace(-4, 3, 120)
i = 0
labels = ['Low Z Geo', 'Low Z Gro', 'Low Z', 'High Z']
colors  = ['grey', 'g', 'r', 'b']
for mu, sig in [(-2.2, .3), (1.5, .3),  (-1, .4), (2, 0.1)]:
    ax.plot(x_values, gaussian(x_values, mu, sig), colors[i], label= labels[i])
    i = i + 1

plt.ylim(0, 1.6)
ax.set_xlabel(r'$\theta$')
ax.legend(loc='upper left')
```
